```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline


from skicaffe import SkiCaffe

import pandas as pd
import os
```

```python
feature_file = "UCM-ResNet-152"
caffe_root = '/usr/local/caffe/'
#model_file = caffe_root + 'models/bvlc_reference_caffenet/bvlc_reference_caffenet.caffemodel'
model_file ='models/models/ResNet-152-model.caffemodel'
arch_file = 'models/models/ResNet-152-deploy.prototxt'
#arch_file = caffe_root + 'models/bvlc_reference_caffenet/deploy.prototxt'
```

```python
src_path = 'UCMerced/Images/'
image_type = (".jpg", ".png", ".JPG", ".PNG", ".JPEG", ".tif", ".tiff", ".TIFF", '.TIF')
 
image_paths = []  
for root, dirs, files in os.walk(src_path):
    image_paths.extend([os.path.join(root, f) for f in files if f.endswith(image_type)])
    
print 'number of images is', len(image_paths)
```

```python
model_prototxt = arch_file
model_trained = model_file
DLmodel = SkiCaffe(caffe_root = caffe_root,model_prototxt_path = model_prototxt, model_trained_path = model_trained, 
                   layer_name = 'fc1000')
```

```python
DLmodel.fit()
print 'Number of layers:', len(DLmodel.layer_sizes)
DLmodel.layer_sizes
```

```python
image_features = DLmodel.transform(X = [image_paths[0]])
image_features.shape
```

```python
bad_images = []
features_dict = {}
for i, image_path in enumerate(image_paths):
    try:
        image_features = DLmodel.transform(X = [image_path])
        features_dict[image_path.split("/")[-1]] = image_features
    except:
        bad_images.append(image_path)
        print i, image_path
```

```python
features_df = pd.DataFrame.from_dict(features_dict, orient='index')
```

```python
features_df.head()
```

```python
features_df.to_csv(feature_file)
```
